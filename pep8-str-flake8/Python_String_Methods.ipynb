## What are the string methods ?

### Split and Join

You can also split it into its single-word components to get a list object of all the words. In fact, the split and join methods make it possible to cut a chain according to a separator to obtain a list, and conversely to reconstruct a chain from a list.

```python
t.split()
```

```python
'abc=:=def=:=ghi=:=jkl'.split('=:=')
```

```python
'=:='.join(['abc', 'def', 'ghi', 'jkl'])
```

Be careful, however, if the separator is a terminator, the result list contains a last empty string. 

```python
'abc;def;ghi;jkl'.split(';') #run again this line by removing the ';' at the end of jlk
```

### Format 

The format method performs a string formatting operation.

```python
'The sum of 1 + 2 is {0}'.format(1+2)
```

```python
# default arguments
print('Hello {}, your balance is {}.'.format('Jimena', 230.2346))

# positional arguments
print('Hello {0}, your balance is {1}.'.format('Jimena', 230.2346))

# keyword arguments
print('Hello {name}, your balance is {blc}.'.format(name='Jimena', blc=230.2346))

# mixed arguments
print('Hello {0}, your balance is {blc}.'.format('Jimena', blc=230.2346))
```

#### Partition method

The partition() method splits the string at the first occurrence of the argument string and returns a tuple containing the part the before separator, argument string and the part after the separator.

```python
string = "Python is fun"

# 'is' separator is found
print(string.partition('is '))

# 'not' separator is not found
print(string.partition('not '))

string2 = "Python is fun, isn't it"

# splits at first occurence of 'is'
print(string2.partition('is'))
```

### ExpandTabs method

The expandtabs() method returns a copy of string with all tab characters '\t' replaced with whitespace characters until the next multiple of tabsize parameter.

The default tabsize is 8. The tab stops are 8, 16 and so on. Hence, there is 5 spaces after 'xyz' and 3 after '12345' when you print the original string.

```python
str = 'xyz\t12345\tabc'
print('Original String:', str)

# tabsize is set to 2
print('Tabsize 2:', str.expandtabs())

# tabsize is set to 3
print('Tabsize 3:', str.expandtabs(3))

# tabsize is set to 4
print('Tabsize 4:', str.expandtabs(4))
```
