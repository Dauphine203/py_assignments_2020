```python
import pandas as pd
import datetime as dt
data = pd.DataFrame.from_dict(json_data["prices"])
data["date"] = data["date"].map(dt.datetime.utcfromtimestamp)
```

```python
import matplotlib.pyplot as plt
import pandas as pd

plt.figure(figsize=(16, 10))

plt.subplot(211)
plt.plot(data["date"], data["close"], "b")
plt.title("Amazon Stock Price")
plt.xlabel("Time")
plt.ylabel("Price")


plt.subplot(212)
plt.title("Google stock trade volume")
plt.hist(data["volume"], bins=50)
plt.ylabel("Volume")
plt.show()
```
